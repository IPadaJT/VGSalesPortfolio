```python
clean_df=initial_df.dropna(how='any')
clean_df
# After dropping data with incomplete rows, we will then make a reviews table and a sales table
```

```python
reviews_df=clean_df[['Name', 'Critic_Score', 'User_Score']].copy()
reviews_df
# This provides a table with all games but scores are split between platforms. We use groupby below to take the mean scores across all platforms that a game is released on. 
```

```python
clean_reviews_df=reviews_df.groupby('Name').mean()
clean_reviews_df
```

```python
sales_df=clean_df[['Name', 'Platform', 'Publisher', 'Developer', 'Total_Shipped', 'Year' ]]
sales_df
# This provides a table with each game's publishing info and sales. But again, some games released on multiple platforms and need to be combined to take average sales across all. 
```

```python
sales_df['Platform_Count']=sales_df.groupby('Name')['Platform'].transform('nunique')
sales_df
# This step adds a new column labeled Platform_Count into the DataFrame that groups by Titles and calculates the number of unique elements for each group back to the original DataFrame
#sales_df.to_csv('test.csv')
```

```python
sales_df['Total_Shipped_Sum']=sales_df.groupby('Name')['Total_Shipped'].transform('sum')
# This step adds a sum of each titles sales by grouping by name and adding all their Total_Shipped values together

sales_df
```

```python
sales_df=sales_df.sort_values(by='Platform_Count', ascending=False)
# This sorts the DataFrame by descending order of the Platform_Count column; i.e. the highest number of platforms first

sales_df
```

```python
sales_df=sales_df.drop_duplicates(subset='Name', keep='first')
# This drops duplicates found within the Title column, specifying to keep 
# only the first occurence which we have set to be the game with the highest platform count

sales_df
```

```python
sales_df=sales_df.drop(columns=['Platform', 'Total_Shipped'])
# Drop the platform and Total_shipped column
```

```python
sales_df.rename(columns={'Total_Shipped_Sum':'Games_Sold'}, inplace=True)
# Renaming sum column to games_sold
```

```python
sales_df.reset_index(drop=True, inplace=True)
# Reset index after dropping rows
sales_df=sales_df.set_index("Name", drop=True)
# Set index to Name column for SQL import
sales_df
```

```python
sorted_sales=sales_df.sort_values(by='Games_Sold', ascending=False)
sorted_sales
```

```python
clean_sales_df=sales_df.copy()
clean_sales_df
```
